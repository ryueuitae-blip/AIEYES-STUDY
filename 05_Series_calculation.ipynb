```python
import pandas as pd
```

```python
#Broadcasting
#반복문을 사용해야 하는 리스트와 달리 연산이 시리즈 객체의 전체 값에 적용됨

from pandas import Series
s = Series([100, 200, 300])

print(s+10) #모든 값에 10을 더하는 연산이 수행됨

```

```python
#사칙연산
#같은 인덱스를 기준으로 사칙연산 적용, 순서 상관 없음

a = Series(data=[100, 200, 300, 400], index=["감자","고구마","마","토란"])
b = Series(data=[1, 2, 3, 4], index=["감자","감자","토란","고구마"])

plus = a + b

print(plus) #결과를 보면 일치하는 인덱스가 없는 경우 NaN를 반환
```

```python
#비교연산
#비교연산의 결과로 논리형 값이 저장된 시리즈를 리턴

s = Series(data = [100,200,300,400])

cond = s>300
print(cond)

```

```python
#필터링
#True/False 값의 리스트를 시리즈에 넘겨주면 True인 값만 필터링한 시리즈를 반환
s = Series(data = [10,20,30,40])
cond =[True,False,False,True] #이 cond는 리스트임..

print(s[cond])

#이걸 비교연산과 결합하면 조건에 따른 필터링 역시 가능하다
cond = s < 25 #이 cond는 시리즈..(인덱스가 동일한 조건 리스트를 만들면 된다.

print(s[cond])
```

```python
# 연습 문제 1
# 고가와 저가 차이 100 이상인 날의 고가를 출력하시오

저가 = Series([10,200,200,400,600])
고가 = Series([100,300,400,500,600])

cond = 고가 - 저가 >= 100

print(고가[cond])
```

```python
# 연습 문제 2
# 종가가 80000원 이상 90000원 미만인 날짜를 출력하시오.

종가 = Series([93000,82400,99100,81000,72300],['05/14','05/15','05/16','05/17','05/18'])

cond = (종가>=80000) & (종가<90000)
#여기서 주의!! panda는 괄호로 비교식을 묶어야 하며, and/or 대신 &/| 사용해야 함(비트연산자 사용)

print(종가[cond].index)
```
