```python
#iloc 속성 이용: 행 번호를 통하여 인덱싱

import pandas as pd

data = [100, 200, 300]
index = ['월', '화', '수']

s = pd.Series(data, index)

s.iloc[0]
print(s)
```

```python
#loc 속성 이용: 인덱스를 통하여 인덱싱

s.loc['화']
```

```python
#대괄호 사용하여 인덱싱 (인덱스를 통하여 인덱싱)

s['월']


#대괄호 사용하여 인덱싱 (행번호를 통하여 인덱싱)
s[1]
```

```python
#iloc 속성 이용: 행 번호 기반 슬라이싱

s.iloc[0:2]
```

```python
#iloc 속성 이용: 인덱스 기반 슬라이싱
#끝 인덱스를 포함함 !!!!!!!!!

s.loc['화':'수']
```

```python
#연속적이지 않은 여러 개의 값을 한 번에 인덱싱


#iloc 속성 이용
target = [0, 2] #가져오고 싶은 데이터의 행 번호를 쉼표로 연결하여 적음
s.iloc[target]
s.iloc[ [0, 2] ]

#loc 속성 이용
target = ['월', '수']
s.loc[target]
s.loc[ ['월', '수'] ]
```
