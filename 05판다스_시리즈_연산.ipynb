```python
# 브로드캐스팅: 연산이 전체 값에 적용
from pandas import Series # 판다스 라이브러리에서 시리즈라는 도구만
s = Series([100, 200, 300]) # 시리즈 객체를 만들고 s에 할
print(s+10)
```

```python
# 사칙연산
high = Series([51500, 51200, 52500, 51500, 51500])
low = Series([50700, 50500, 50500, 50800, 50700])
diff = high - low
print(diff)
```

```python
high = Series(data= [51500, 51200, 52500], index= ["5/1","5/2", "5/3" ])
low = Series(data= [50700, 50500, 50500 ], index = ["5/1", "5/2", "5/4"])
diff = high - low
print(diff)
```

```python
# 시리즈 비교연산
s = Series(data= [100,200,300,400,500])
cond= s> 300
print(cond)
```

```python
# 시리즈 필터링
from pandas import Series
s = Series(data= [100,200,300,400,500])
cond=[False,False,False,True,True]
s[cond]
```

```python
s = Series(data= [100,200,300,400,500])
cond= s> 300
print(s[cond])
```

```python
import pandas as pd
저가 = pd.Series([100,200,300,400,600])
고가 = pd.Series([100,300,400,500,600])

cond = (고가 - 저가) >= 100
고가[cond]
```

```python
data = [93000, 82400, 99100, 81000, 72300]
index = ['05/14', '05/15', '05/16', '05/17', '05/18']
종가 = pd.Series(data=data, index=index)
print(종가)
```

```python
조건 = (종가 >= 80000) & (종가 <= 90000)
print(종가[조건])
```

```python
종가[조건].index
```

```python
for date in 종가[조건].index:
    print(date)
```
