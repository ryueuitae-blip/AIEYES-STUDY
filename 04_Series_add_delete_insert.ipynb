```python
#Series 값 추가(딕셔너리와 유사)
import pandas as pd

data = [100,200,300]
index = ["월","화","수"]
s = pd.Series(data,index)

s.loc["목"] = 400 #인덱스 기준으로 값을 할당
# s.iloc[500] = "금" <- 이건 안된다...

print(s)
```

```python
#Series  삭제
#원본은 유지, 값이 삭제된 새로운 시리즈 객체를 리턴

s1 = s.drop("월")
s2 = s.drop(["화","목"]) #여러 값도 OK!

print(s1) #화,수,목만 출력
print()
print(s2) #월,수만 출력
print()
print(s) #원본은 월,화,수,목 출력
```

```python
#Series 수정
#행 번호 or 인덱스를 활용하여 value 수정

#행 번호->iloc, 인덱스->loc
s.iloc[0] = 1000
s.loc["화"] = 2000

#그냥 대괄호로 쓰면 인덱스 -> 행 번호 순으로 일치값에 대해 수정사항 반영
s[2] = 500


print(s)
```
