```python
#iloc를 활용한 슬라이싱
#시작 행번호 : 끝 행번호
data = [100,200,300]
index = ["월","화","수"]
s = Series(data,index)
print ( s.iloc[0:2]) # 0번째, 첫번째 가져옴 마지막 포함 안됨!!
```

```python
#loc를 활용한 슬라이싱
#iloc와의 차이 == 시작 인덱스 : 끝 인덱스가 포함이 됨!!!
data = [100,200,300]
index = ["월","화","수"]
s = Series(data,index)
print ( s.loc["월":"화"])
```

```python
#여러 값 인덱싱 / 멀티인덱싱
#파이썬과 달리 연속적이지 않은데 여러값을 가져올수있음 
#iloc loc 둘다 가능 
#[0,2]
#0번과 2번을 가져오겠다는 뜻 

s.iloc[[0,2]] #[]안에 다시 [] 사용 
```

```python
s.loc[["월","수"]] #iloc, loc 사용 권장 
```
