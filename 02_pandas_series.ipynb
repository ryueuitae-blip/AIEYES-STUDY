```python
#Series 객체 생성

import pandas as pd

arr = [0,1,2,3]
arr
```

```python
s = pd.Series(arr) #arr라는 리스트를 인자값으로 하는 Series 객체를 s에 할당
s
```

```python
type(s) #타입은 Series
```

```python
#Series 내부 구조
#행 번호와 인덱스를 사용해 데이터 관리(접근)
#행 번호: 자동으로 부여됨(수정 불가) -> 별도로 보이는 것은 X
#인덱스: 문자/숫자로 설정 가능(default:0부터 시작하는 정수), 딕셔너리의 키와 유사 개념
#행 번호는 앞/마지막 값 가져올 때 편리 ex) [-1] -> 뒤에서 첫번째
#인덱스는 명시적인 경우 그 이름으로 값 가져올 때 편리 ex) '류희수'
```

```python
data = [100, 200, 300]
index = ['월', '화', '수']

s = pd.Series(data,index) #인덱스와 값을 가지고 Series 객체 생성
s
```

```python
#Series 클래스의 속성
#index: 시리즈 객체의 인덱스에 접근 ex) (객체명).index
#values: 시리즈 객체의 값에 접근 ex) (객체명).values
```

```python
s.index
```

```python
s.values
```
