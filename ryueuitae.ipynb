```python
# Numpy: 행렬,다차원 연산
import numpy as np
#Numpy라는 모듈을 불러온다. 이제부터 Numpy를 np라 부른다
```

```python
data = [1,2,3,4]
result= []

for i in data: #데이터 리스트 안에 i에 대한 반복
  result.append(i*10) # result라는 리스트에 i에 10을 곱한 값을 추가해라

print(result)
```
