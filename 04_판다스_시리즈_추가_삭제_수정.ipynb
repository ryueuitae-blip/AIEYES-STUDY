```python
# Series에 값 추가
# s.loc[인덱스] = 값

from pandas import Series

data = [100, 200, 300]
index = ["월", "화", "수"]
s = Series(data, index)

s.loc["목"] = 400
s["목"] = 400
print(s)
```

```python
print(s)
```

```python
# Series 삭제
# drop 메서드: 원본 유지, 값이 삭제된 시리즈 객체 리턴
# 사용법: s.drop('인덱스'), s.drop(['인덱스1', '인덱스2']

s1 = s.drop('수') # 단일 값: 대괄호 필요X
print(s1)
s2 = s.drop(['월', '수']) # 여러 값: 리스트로 묶어야 함 -> 대괄호 사용
print(s2)
print(s)

# 원본 변경법: 원본 시리즈 객체에 drop이 바로 적용됨

s.drop('목', inplace=True)
print(s)
```

```python
# Series 수정: 행번호 or 인덱스 사용 -> 값(calue) 수정

s.iloc[0] = 3333 # 행번호 사용(iloc)
print(s)

s.loc["수"] = 576 # 인덱스 사용(loc)
print(s)

s["화"] = 2000 # 찾는 순서: 1. 인덱스, 2. 행번호
```
