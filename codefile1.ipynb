```python
#Numpy: 수치계산 효율 목적 라이브러리
import Numpy as np # Numpy 불러오기, 이제 np로 부름
```

```python
data = [1,2,3,4]
result= []

for i in data: # data 리스트 안의 i에 대한 반복문
  result.append(i*10) # result 라는 리스트에 i에 10을 곱해서 들어간다

print(result)
```
