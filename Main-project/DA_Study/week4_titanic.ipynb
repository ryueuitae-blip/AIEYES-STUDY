```python
import numpy as np
import pandas as pd
import seaborn as sns #seaborn안에 타이타닉 데이터셋이 있음! 
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head() #첫 5개 데이터 확인
```

```python
titanic.describe() #어떤식의 데이터분포를 가지는지 확인가능 
```

```python
#누락값 드랍시키기
titanic.dropna().describe() 
```

```python
titanic.var(numeric_only=True)
#함수 안에 숫자로만 계산하라는 옵션을 명시
```

```python
#(Mean Absolute Deviation, 평균 절대 편차) 함수는 Pandas 버전 1.5.0부터 제거(deprecated)되어 더 이상 사용할 수 없다한다...
# MAD는 데이터들이 얼마나 흩어졌는지 알려주는데, 예를 들어 승객들의 나이는 11살정도 차이나니깐 매우 다양하다는걸 알수있음

# 1. 숫자형 데이터만 선택
numeric_titanic = titanic.select_dtypes(include=['number'])

# 2. mad 직접 구현: (데이터 - 평균)의 절대값의 평균
titanic_mad = (numeric_titanic - numeric_titanic.mean()).abs().mean()

print(titanic_mad)
```

```python
titanic.groupby('class').count()
#클래스별로 몇명이 탔는지 궁금해 
#3등급 생존자가 제일 많아보이지만 아님! 
```

```python
# 등급별 생존자 '수' 직접 확인하기
titanic.groupby('class')['survived'].sum()
# 1등급 216명중 136명 생존
# 3등급 491명중 119명 생존... 
```

```python
# 클래스별 생존율 계산
survival_by_class = titanic.groupby('pclass')['survived'].mean()
print(survival_by_class)
# 3등급은 24퍼센트밖에 안된다....
```
