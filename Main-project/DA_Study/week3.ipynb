```python
import pandas as pd 

#첫번째 방법
# iloc = integer location 행 번호를 가지고 인덱싱 하는 방법

data = [100,200,300]
index = ["월","화","수"]
s = pd.Series(data,index)
```

```python
s.iloc[0]
```

```python
s.iloc[2]
```

```python
s.iloc[-1] #함수가 아니라 () 사용시 에러 발생
```

```python
# 두번째 방밥: 인덱스를 사용하는 방법
# loc = location

print(s.loc["월"]) 
```

```python
# 세번째 방법: [] 사용해서 인덱스를 통해서 가져오는 방법
# 잘 사용하지는 않는다고 함
# 다른것들을 명시적으로 쓰는걸 권장
# 예전 버전에서는 에러 발생함 

s[0]
```

```python
# 슬라이싱 
# iloc[시작 행:끝 행]
# 파이썬의 리스트처럼 

print(s.iloc[0:2])
```

```python
# 두번째 슬라이싱
# loc[시작 인덱스: 끝 인덱스
# 끝 인덱스가 포함된다는 특징

print(s.loc["화":"수"])
```

```python
# 여러값 인덱싱
# 파이썬은 구간, 특정값만 가능한데,
# 동시에 선택적으로 또는 비연속적인 여러개의 값을 인덱싱 가능하다는게 장점!

# iloc vs loc
# 가져오고 싶은 번호를 적기 

m = [1,2,3,4]

m[[0,2]] # 파이썬에서는 존재하지않음
```

```python
# 시리즈 객체에 추가, 삭제, 수정
```

```python
# 값 추가

data = [100,200,300]
index = ["월","화","수"]
s = pd.Series(data,index)

s.loc["목"] = 400 # s["목"] =400 과 동일 
print(s)
```

```python
# 값 삭제
# 원본은 유지되고, 새로운 객체로 리턴
s1 = s.drop("월") # ()사용, 여러개 가능 
print(s1)
print(s)
```

```python
# 값 수정 

s.iloc[0] = 1000 
s.loc["수"] = 3000
s
```

```python
s.drop('목', inplace=True)
#원본에 바로 삭제하고싶으면 inplace=True 사용
s
```

```python
s['수'] = 3000
s
```

```python
s.drop(['수','화'],inplace =True) #응용한다면 여러개 삭제도 가능
s
```

```python
# 5주차 연산
# 기존 엑셀에서는 반복문을 사용했음
# 판다스는 반복문 없이도 모든 데이터를 수정함 = 브로드캐시팅

s = pd.Series([100,200,300])
print(s +10)
```

```python
# 사칙연산
# how 와 low로 나뉘어짐
high = pd.Series([51500,51200,52500,51500,51500])
low = pd.Series([50700,50500,50500,50580,50700])
diff = high - low # 두개의 시리즈 객체에서 동일한 인덱스를 가진 값들을 바꿔줌

print(diff)
```

```python
# 다른 인덱스를 갖는 경우

high = pd.Series([51500,51200,52500], index=["5/1","5/2","5/3"])
low = pd.Series([50700,50500,50500], index=["5/1","5/2","5/4"])
diff = high - low
print(diff)

# 판다스에서는 데이터의 개수와 인덱스의 개수가 반드시 일치해야함
# 순서는 상관없지만 일치하지 않으면 NaN 으로 존재하지 않는다고 뜸 
```

```python
# Boolean 타입

s = pd.Series(data=[100,200,300,400,500])
cond = s>300
print(cond)
```

```python
# 필터링 

s = pd.Series(data=[100,200,300,400,500])
cond = [False,False,False,True,True] #리스트에 T/F가 들어가있는 상태 
s[cond]

# 불리언 리스트는 값이 true인거만 필터링한다
```

```python
시가 = pd.Series([100,200,300,400,600])
종가 = pd.Series([100,300,400,500,600])

조건 = (시가 >= 300)
print(종가[조건]) #시가가 300이상의 종가 데이터를 가져오는 예시 
```

```python
저가 = pd.Series([10,200,200,400,600])
고가 = pd.Series([100,300,400,500,600])

cond = (고가 - 저가) >= 100
고가[cond] # cond = condition 조건 
```

```python
data = [93000,82400,99100,81000,72300]
index = ['5/14','5/15','5/16','5/17','5/18']

종가 = pd.Series(data=data , index=index)
종가 #생성하고 
```

```python
조건 = (종가 >= 80000) & (종가 < 90000) # 80000원 이상, 90000 미만으로 설정 
#and 대신 &
#or 대신 |
#not 대신 ~
종가[조건]
#날짜만 가져올려면?
종가[조건].index 
```

```python
for date in 종가[조건].index:
    print(date)
#이렇게 반복문으로도 가져올 수 있음! 
```
