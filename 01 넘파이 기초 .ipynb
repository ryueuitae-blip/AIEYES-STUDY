```python
# numpy: 행렬, 다차원 배열 연산
import numpy as np # numpy 모듈 불러오기, 이제 np로 부름
```

```python
data = [1,2,3,4]
result= []

for i in data: # data 리스트 안의 i에 대한 반복문
  result.append(i*10) # result 라는 리스트에 i에 10을 곱해서 들어간다

print(result)
```

```python
import numpy as np
arr = np.array([1,2,3,4]) # list 타입을 다차원 배열로
arr10 = arr * 10 # 위에서와 달리 반복문 필요X

print(arr10)
```
