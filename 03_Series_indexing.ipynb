```python
import pandas as pd
data = [100,200,300]
index = ["월","화","수"]
s = pd.Series(data, index)
s
```

```python
#행 번호로 접근
s.iloc[0],s.iloc[1],s.iloc[2],s.iloc[-1]
```

```python
#인덱스로 접근
s.loc["월"]
```

```python
#요즘은 그냥 대괄호만 사용해도 됩니다~
s["화"]
s[1] #그러나 권장하지 않음(이전 버전과의 호환성 문제. 경고문 참고)
```

```python
#행 번호를 사용한 슬라이싱
print(s.iloc[0:2])
```

```python
#인덱스를 사용한 슬라이싱
print(s.loc["월":"화"]) #행 번호와 달리 명시한 끝 값까지 가져옴
```

```python
#여러 값 인덱싱(타겟팅)
target = [0,2] #가져올 값을 타겟팅한 리스트(행 번호)
target2 = ["월","수"] #타겟팅 리스트(인덱스)

print(s.iloc[target])
print(s.loc[target2])
```
