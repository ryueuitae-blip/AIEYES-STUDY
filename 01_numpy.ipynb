```python
#임포트(with 별칭)
import numpy as np
```

```python
#파이썬 리스트는...
data = [1,2,3,4]
data * 10 #값에 곱연산을 수행하는게 아니라 반복으로 인식
```

```python
#반복문을 활용하면 해결은 가능.. 하지만 불편함

data = [1,2,3,4]
result = []

for i in data:
    result.append(i * 10)

print(result)
```

```python
#numpy를 사용하여 배열 arr 생성
arr = np.array(data)
```

```python
arr
```

```python
type(arr) #리스트에서 ndarray라는 타입으로 변경된 것을 확인
```

```python
arr * 10 #리스트와 달리 값의 배수로 인식 -> 다차원 행렬, 배열 처리에 용이
```

```python
#2차원 데이터 다루기
#리스트의 한계 -> 열 단위 접근 불가(행 단위만 가능, 리스트는 범용적 목적이므로 transpose 등 행렬 특화 연산 지원 X)

data = [
    [1,2,3,4],
    [10,20,30,40]
]

print(data[1]) # 행 단위...
print(data[1][3]) # 요소 단위...
#print(data[][1]) <- 열 단위? 이런건 당연히 안된다
```
